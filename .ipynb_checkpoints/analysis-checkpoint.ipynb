```python
import pandas as pd
import numpy as np
import chardet
```

```python
file_path = "98-401-X2021003_English_CSV_data.csv"


# Read the CSV file with 'utf-8' encoding and handle errors with 'latin-1'
try:
    df = pd.read_csv(file_path, encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv(file_path, encoding='latin-1')
```

```python
columns_to_keep = ['ALT_GEO_CODE','GEO_NAME', 'CHARACTERISTIC_NAME','C1_COUNT_TOTAL', 'C2_COUNT_MEN+', 'C3_COUNT_WOMEN+']
```

```python
df =df[columns_to_keep]
```

```python
df.tail(25)
```

```python
df['Level'] = (df['CHARACTERISTIC_NAME'].apply(len)-df['CHARACTERISTIC_NAME'].str.strip().apply(len))/2
```

```python
max_level = int(df['Level'].max())
```

```python
level_val=df['Level'].unique()
```

```python
level_list =[]
```

```python
level_list = ["Lvl-" + str(i) for i in range(0, max_level + 1)]
```

```python
level_list
```

```python
for column_name in level_list:
        df[column_name] = column_name
```

```python
for column_name in level_list:
        df[column_name] = np.nan
```

```python
df
```

```python
df_play = df.copy()
```

```python
# Iterate through the DataFrame rows
for index, row in df_play.iterrows():

    for level_column in level_list:
            if row['Level'] == int(level_column.split('-')[1]):
                characteristic_name = row['CHARACTERISTIC_NAME'].strip()
                df_play.at[index, level_column] = characteristic_name
```

```python
# iterate each level column and update the NaN ones based on the hierarchy
level_val=df['Level'].unique()
for i in level_val:
    i = int(i)
    last_characteristic_name = ''
    # Iterate through the DataFrame rows
    for index, row in df_play.iterrows():
        if pd.notna(row[level_list[i]]):
            last_characteristic_name = row['CHARACTERISTIC_NAME'].strip()
        elif row['Level'] >= i:
            # Update the 'Lvl-x' column with the last_characteristic_name until the next 0 occurs
            df_play.at[index, level_list[i]] = last_characteristic_name
```

```python
df_play.to_csv("Output/census_data_with_levels.csv", index=False)
```

## Seperator

```python
#add the location of the file
df_census = pd.read_csv("Output/census_data_with_levels.csv", sep=',') 
df_census.head(10)
```

```python
df_census['Lvl-0'].unique()
```

```python
vis_minority_level = df_census[df_census['Lvl-0']=='Total - Visible minority for the population in private households - 25% sample data']
pivot_vis_minority = vis_minority_level.pivot_table(
    index=['ALT_GEO_CODE', 'GEO_NAME'],
    columns='Lvl-1',
    values=['C1_COUNT_TOTAL', 'C2_COUNT_MEN+', 'C3_COUNT_WOMEN+']
)
pivot_vis_minority.shape
```

```python
def create_pivot_table(df, filter_condition):
    # Filter the DataFrame based on the condition
    filtered_df = df[df['Lvl-0'] == filter_condition]

    # Create the pivot table
    pivot_table = filtered_df.pivot_table(
        index=['ALT_GEO_CODE', 'GEO_NAME'],
        columns='Lvl-1',
        values=['C1_COUNT_TOTAL', 'C2_COUNT_MEN+', 'C3_COUNT_WOMEN+']
    )

    return pivot_table

# usage:
pivot_vis_minority = create_pivot_table(df_census, 'Total - Visible minority for the population in private households - 25% sample data')
pivot_vis_minority.shape
```

```python
pivot_indigen = create_pivot_table(df_census, 'Total - Indigenous ancestry for the population in private households - 25% sample data')
pivot_indigen.shape
```

```python
pivot_citiz = create_pivot_table(df_census, 'Total - Citizenship for the population in private households - 25% sample data')
pivot_citiz.sample(5)
```

```python
pivot_immmig = create_pivot_table(df_census, 'Total - Immigrant status and period of immigration for the population in private households - 25% sample data')
pivot_immmig.shape
```

```python
pivot_immmigage = create_pivot_table(df_census, 'Total - Age at immigration for the immigrant population in private households - 25% sample data')
pivot_immmigage.shape
```

```python
pivot_immmigplaceofbirth = create_pivot_table(df_census, 'Total - Place of birth for the immigrant population in private households - 25% sample data')
pivot_immmigplaceofbirth.shape
```

```python
pivot_immmigethnic = create_pivot_table(df_census, 'Total - Ethnic or cultural origin for the population in private households - 25% sample data')
pivot_immmigethnic.shape
```

```python
# Assuming these are your DataFrames
dataframes = [
    pivot_vis_minority,
    pivot_indigen,
    pivot_citiz,
    pivot_immmig,
    pivot_immmigage,
    pivot_immmigplaceofbirth
]

# Merge DataFrames iteratively
merged_df = dataframes[0]
for df in dataframes[1:]:
    merged_df = merged_df.merge(df, on=["GEO_NAME"], how="outer")  # Using outer join for completeness

```
